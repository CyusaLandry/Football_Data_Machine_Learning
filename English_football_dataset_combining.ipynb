import pandas,numpy and os which will help find the folder of my csv files and combine them together

```python
import pandas as pd
import numpy as np
import os 
folder_path = 'D:\Big Data\Final Exam\Final folder\datasets' 
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
csv_files.sort()
```

Combine all csvs into a single dataset

```python
combined_df = pd.concat(
    [pd.read_csv(os.path.join(folder_path, file)) for file in csv_files],
    ignore_index=True
)

# Save the result to a new CSV file
combined_df.to_csv('english_football_data.csv', index=False)

print(f" Combined {len(csv_files)} files into 'combined_output.csv'")
combined_df.head() 
```
